```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243) ## initialize the random state
```

```python
# specifying mean and covariance for dummy class1 and class2 data

mean1 = np.array([0,0,0])
cov1 = np.array([[1,0,0],[0,1,0],[0,0,1]])

class1 = np.random.multivariate_normal(mean1,cov1,100)

mean2 = np.array([1,1,1])
class2 = np.random.multivariate_normal(mean2,cov1,100)
```

```python
# Plotting a 3D graph

from mpl_toolkits.mplot3d import Axes3D, proj3d

fig = plt.figure(figsize=(8,8)) # fig size
ax = fig.add_subplot(111,projection='3d') # 111 : 1-> Rows: 1-> Columns:1->Position
ax.plot(class1[:, 0], class1[:, 1], class1[:, 2], 'o')
ax.plot(class2[:, 0], class2[:, 1], class2[:, 2], '^')
plt.show()
```

```python
data_combine = np.concatenate((class1,class2))
data_combine.shape
```

```python
pca = PCA()
class_trans = pca.fit_transform(data_combine)

fig = plt.figure(figsize=(8,8)) # fig size
ax = fig.add_subplot(111,projection='3d') # 111 : 1-> Rows: 1-> Columns:1->Position
ax.plot(class_trans[0:100, 0], class_trans[0:100, 1], class_trans[0:100, 2],'o') # first 100 enteries are of class1
ax.plot(class_trans[100:200, 0], class_trans[100:200, 1], class_trans[100:200, 2], '^') # next hundred rows are of class2
plt.show()
```

# Note : Even if we keep the dimensions same , we get a slight different plot because PCA has changed our data values slightly.


1. PCA tries to maximise values(maximises useful data present) in some directions instead of keeping useful data in all dimensions. 
2. Dimensions(components) having most variance are most important.

```python
pca.components_
```

```python
pca.explained_variance_ratio_ 
```

# Converting to 2D :

```python
pca1 = PCA(n_components = 2)
class_trans1 = pca1.fit_transform(data_combine)
class_trans1.shape
```

```python
pca1.components_ # the two components considered here are same as top two of pca considered above(with dimensions = 3) 
```

```python
pca1.explained_variance_ratio_
```

```python
# Plotting the 2D plot
# the data still seems to be separated
plt.plot(class_trans1[0:100, 0], class_trans1[0:100, 1],'o') # first 100 enteries are of class1
plt.plot(class_trans1[100:200, 0], class_trans1[100:200, 1], '^') # next hundred rows are of class2

plt.show()
```

```python
# getting the original data from pca converted data
data_approx = pca1.inverse_transform(class_trans1)
print(data_approx.shape)

# plotting the approx data we get 
# Note that the data seems to be in 3d , but actually it all lies in a 2d Plane

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection = '3d')

ax.plot(data_approx[0:100,0],data_approx[0:100,1],data_approx[0:100,2],'o')
ax.plot(data_approx[100:200,0],data_approx[100:200,1],data_approx[100:200,2],'^')

plt.show()
```
